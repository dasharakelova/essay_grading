# Предобратока данных
**Получение _data.json_ и _data.csv_**

```python
from json import loads, dumps
import os
from collections import defaultdict

os.chdir("data")

stats = defaultdict(lambda: [0, 0])

inf = {}

for substr in os.listdir("."):
    if "exp" in substr:
        with open(substr) as f:
            data = loads(f.read())
            if data["meta"]["subject"] == "rus":
                stats[data["meta"]["id"]][0] += sum(map(int, data["criteria"].values()))
                stats[data["meta"]["id"]][1] += 1
                inf[data["meta"]["id"]] = {
                    "text": data["text"],
                    "year": data["meta"]["year"],
                }
os.chdir("..")
```

```python
with open("output.txt", "w") as out:
    for substr in stats.items():
        print(*substr, file=out)
```

```python
for substr in stats.keys():
    inf[substr]["score"] = stats[substr][0] / stats[substr][1]
```

```python
with open("data.json", "w") as f:
    print(dumps(inf, ensure_ascii=False, indent=4), file=f)
```

```python
vals = [i["score"] for i in inf.values()]
print(sum(vals) / len(vals), max(vals))
```

```python
sorted(vals, reverse=True)
```

```python
from csv import writer

with open("data.csv", "w") as f:
    w = writer(f)
    w.writerow(["id", "score", "year", "text"])
    for substr in inf.items():
        w.writerow([substr[0], substr[1]["score"], substr[1]["year"], substr[1]["text"]])
```
