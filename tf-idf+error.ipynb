```python
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
import json
with open('output.json') as cat_file:
    data = json.load(cat_file)
```

```python
n=0
k =0
for key, value in data.items():
    n += value['score']
    k += 1
middle = n/k
```

```python
import nltk
import numpy as np
import random
import string

import bs4 as bs
import urllib.request
import re
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression

import os
```

```python
X_train = []
y_train = []
for key, value in data.items():
    X_train.append(value['text'])
    y_train.append(value['score'])
X_test = X_train[-600:]
X_train = X_train[:-600]
y_test = y_train[-600:]
y_train = y_train[:-600]
print(y_train)
```

```python
y_random = []
for i in range(len(y_test)):
    y_random.append(random.uniform(min(y_train), max(y_train)))
print(y_random)
```

```python
y_norm = []
for i in range(len(y_test)):
    y_norm.append(np.random.normal(loc=middle))
print(y_norm)
```

```python
y_mid = []
for i in range(len(y_test)):
    y_mid.append(np.random.normal(loc=middle))
print(y_mid)
```

```python
vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)
# print(y_pred)
```

```python
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
print("Mean squared error random: %.2f" % mean_squared_error(y_test, y_random))
print("Mean squared error norm: %.2f" % mean_squared_error(y_test, y_norm))
print("Mean squared error mid: %.2f" % mean_squared_error(y_test, y_mid))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
```
