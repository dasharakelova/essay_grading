```python
from csv import DictReader

with open("data.csv") as dt:
    data = list(DictReader(dt))

print(*data[:5], sep="\n")
```

Запускать этот блок, для лемматизации

```python
from pymorphy3 import MorphAnalyzer
from razdel import tokenize

morph = MorphAnalyzer()

dt = [0] * len(data)

for i in range(len(data)):
    tkn = list(tokenize(data[i]["text"]))
    dt[i] = [0] * len(tkn)
    for j in range(len(tkn)):
        dt[i][j] = morph.parse(tkn[j].text)[0].normal_form

print(*dt[:5], sep="\n")
```

Запускать этот блок, для отсутствия лемматизации

```python
from pymorphy3 import MorphAnalyzer
from razdel import tokenize

morph = MorphAnalyzer()

dt = [0] * len(data)

for i in range(len(data)):
    tkn = list(tokenize(data[i]["text"]))
    dt[i] = [0] * len(tkn)
    for j in range(len(tkn)):
        dt[i][j] = tkn[j].text

print(*dt[:5], sep="\n")
```

```python
for i in range(len(dt)):
    dt[i] = " ".join(map(lambda x: x.lower(), dt[i]))

print(*dt[:5], sep="\n")
```

**0.8** - доля датасета, уходящая на обучение модели

```python
border = int(len(dt) * 0.8)
corpus = dt[:border]
corpus_ans = [float(data[i]["score"]) for i in range(border)]
test = dt[border:]
test_ans = [float(data[i]["score"]) for i in range(border, len(data))]

print(*corpus[:5], sep="\n")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(ngram_range=(1, 3))
train_inp = vectorizer.fit_transform(corpus)
test_inp = vectorizer.transform(test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

regr = LinearRegression()

regr.fit(train_inp, corpus_ans)

pred = regr.predict(test_inp)

print("Coefficients: \n", regr.coef_)
print("Mean squared error: %.2f" % mean_squared_error(test_ans, pred))
print("Coefficient of determination: %.2f" % r2_score(test_ans, pred))

print(f"Max coeff: {max(regr.coef_)}")
print(f"Min coeff: {min(regr.coef_)}")
print(f"Number of coeffs: {len(regr.coef_)}")
```
